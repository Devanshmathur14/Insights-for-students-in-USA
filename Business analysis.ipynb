### 1.	Do universities with a high number of applications are the preferred ones by students? in other words, could the number of applications tell us that a university is one of the most preferred by students?

This can be deduced by finding the difference between total applicants and actual admissions. Higher the difference means more people prefer that university.

```python
df1['diff_applicant_and_admissions'] = df1['Applicants total'] - df1['Admissions total']
```

```python
df2 = df1.sort_values(['diff_applicant_and_admissions'], ascending=False)
```

```python
df3 = df2.head(10)
```

```python
plt.figure(figsize = (12,5), dpi = 200)
sns.barplot(df3['Name'], df3['diff_applicant_and_admissions'])
plt.xticks(rotation = 90)
```

```python
dfsf = pd.DataFrame(data=sf.index, columns=['State abbreviation'])
dfv = pd.DataFrame(data=sf.values, columns=['Total admissions'])
df3 = pd.merge(dfsf, dfv, left_index=True, right_index=True)
```

```python
df4 = pd.merge(df3, df_sgdp, left_index=True, right_index=True)
```

```python
df4.sort_values(by = 'Total admissions', ascending = False , inplace = True)
```

```python
df4
```

```python
plt.figure(figsize = (12,5), dpi = 200)
sns.barplot(df4['State abbreviation'],df4['Total admissions'] )
plt.xticks(rotation = 90)
```

```python
plt.figure(figsize = (12,5), dpi = 200)
sns.barplot(df4['State abbreviation'],df4['gdp_million'] )
plt.xticks(rotation = 90)
```

##### From above 2 graphs we ca clearly see that total number of applicants is higher in states like California, New York, Texas, Illinois, Florida, Michigan which has highest gdp which signfies students prefer a university because it belongs to a state with a high GDP per capita.
#### Also Population of these states is highest among all other states so that signifies that students prefer universities from highly populated states. 

### 8.	Do students prefer a university based on the possibility of a higher, additional academic degree in the same university?

```python
df2['Highest degree offered'].unique()
```

```python
df2['Highest degree offered'] = df2['Highest degree offered'].map({"Doctor's degree - research/scholarship and professional practice": "Doctor",
                                   "Doctor's degree - other" : "Doctor",
                                   "Doctor's degree - research/scholarship" : "Doctor",
                                   "Master's degree" : "Master",
                                   "Bachelor's degree" : "Bachelor",
                                   "Doctor's degree - professional practice" : "Doctor"})
```

```python
df2['Highest degree offered'].value_counts()
```

```python
df2.groupby(by = 'Highest degree offered')['Applicants total'].mean()
```

##### Here we can clearly see that average number of applicants is sigificantly larger in Universities that provide Doctor's degree as highest level of education followed by Master's and least preferred universities are the one which have Bachelor as there highest levelof Education. So, we ca cofidently say that students prefer a university based on the possibility of a higher, additional academic degree in the same university.
